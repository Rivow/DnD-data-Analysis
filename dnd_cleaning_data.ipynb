```python
def find_max_level(levels):
    maxl = 0
    clas = ''
    for lvl in levels:
        split = lvl.split()
        l = int(split[-1])
        if l > maxl:
            maxl = l
            clas = ' '.join(split[0:-1])
        elif l == maxl:
            clas = 'Equal Multiclass'
    return clas
```

```python
multi = pd.DataFrame(np.zeros((len(data), 1)))
multi.columns = ['multiclass']
multi
```

```python
dndclass = data['class'].str.split('|')
n = 0
for index, value in dndclass.items():
    dndclass[index] = find_max_level(value)
    if len(value) > 1:
        multi['multiclass'][index] = True
    else:
        multi['multiclass'][index] = False
data = data.merge(dndclass.to_frame(), left_index=True, right_index=True, suffixes=('', '_main'))
data = data.merge(multi, left_index=True, right_index=True)
data.head()
```

```python
sub_class = data['subclass']
sub= sub_class.str.split('|').dropna()
jclass = data['justClass'].str.split('|').dropna()
for index, value in sub.items():
    if type(value) == list:
        ls = []
        for v in value:
            if v:
                ls.append(v)
        sub[index] = ls
```

**Crete a data frame with count of the subclasses**

```python
sub_dict = {}

for index, value in sub.items():
    if len(value) > 1:
        for v in value:
            if v in sub_dict:
                sub_dict[v] += 1
            else:
                sub_dict[v] = 1

    else:
        if value:
            if value[0] in sub_dict:
                sub_dict[value[0]] += 1
            else:
                sub_dict[value[0]] = 1
                    
hexb = sub_dict.pop('Hexblade')
forge = sub_dict.pop('Forge')
# since I found out they where doubled removed and added to the original
sub_dict['The Hexblade'] += hexb
sub_dict['Forge Domain'] += forge
```

**Create Class-Subclass data frame to know wich subclases go with wich class**

```python
sub_class_dict = {}
k = sub_dict.keys()

for index, value in sub.items():
    if len(value) == 1:
        if value[0] not in sub_class_dict:
            sub_class_dict[value[0]] = data['class_main'][index]

# since I found out they where doubled removed                
sub_class_dict.pop('Hexblade')
sub_class_dict.pop('Forge')
```

**Now merge both dicts to a Data Frame**

```python
class_sub = {'sub_class': [x for x in sub_class_dict.keys()], 'class': [x[1] for x in sub_class_dict.items()]}
count_sub =  {'sub_class': [x for x in sub_dict.keys()], 'count': [x[1] for x in sub_dict.items()]}
sub = pd.DataFrame.from_dict(class_sub)
count = pd.DataFrame.from_dict(count_sub)
sub_classes_df = sub.merge(count, on=['sub_class'])
```

**creating a feat count to analyse popularity**

```python
# Dict all oficial D&D and count of feats used per class classes for feat analysis
off_class = ['Artificer', 'Barbarian', 'Bard', 'Cleric', 'Druid', 'Fighter', 'Monk', 
             'Paladin', 'Ranger', 'Rogue', 'Sorcerer', 'Warlock', 'Wizard']

feat_dict = {}
feats = data['feats'].str.split('|').dropna()

for index, value in feats.items():
    
    if len(value) > 1:
        for v in value:
            if v in feat_dict:
                feat_dict[v] += 1
            else:
                feat_dict[v] = 1
    else:
        if value[0] in feat_dict:
            feat_dict[value[0]] += 1
        else:
            feat_dict[value[0]] = 1
```

```python
feat_class = {}
feats_list = [f for f, _ in feat_dict.items()]
final_dict = {'feat': [f for f, _ in feat_dict.items()], 'count': [c for _ ,c in feat_dict.items()]}

# Get amount of each feat used per Class
for cl in off_class:
    temp = [0 for _ in feats_list]
    for index, value in feats.items():
        if len(value) > 1:
            for v in value:
                if data['class_main'][index] == cl:
                    temp[feats_list.index(v)] += 1
        else:
            if data['class_main'][index] == cl:
                temp[feats_list.index(value[0])] += 1
    
    final_dict[cl] = temp
        
feats = pd.DataFrame.from_dict(final_dict)
feats
```
