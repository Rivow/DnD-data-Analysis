```python
import pandas as pd
```

```python
data = pd.read_csv('dnd_chars_unique.tsv', sep='\t')
```

```python
data.head(2)
```

```python
def find_max_level(levels):
    maxl = 0
    clas = ''
    for lvl in levels:
        split = lvl.split()
        l = int(split[-1])
        if l > maxl:
            maxl = l
            clas = split[0]
        elif l == maxl:
            clas = 'Equal Multiclass'
    return clas
```

```python
dndclass = data['class'].str.split('|')
n = 0
for index, value in dndclass.items():
    dndclass[index] = find_max_level(value)
data = data.merge(dndclass.to_frame(), left_index=True, right_index=True, suffixes=('', '_main'))
data.head()
```

```python
data.to_csv('clean_dnd_chars_unique.csv')
```
